```python
import pandas as pd
```

```python
bnc=pd.read_csv('/content/BinanceCoin.csv')
btc=pd.read_csv('/content/Bitcoin.csv')
eth=pd.read_csv('/content/Ethereum.csv')

bnc.head()
```

```python
# Clean column names---- Standardization
def clean_col(df):
  df.columns = (df.columns
                .str.strip()
                .str.lower()
                .str.replace(" ","_"))
  return df
bnc=clean_col(bnc)
btc=clean_col(btc)
eth=clean_col(eth)
```

```python
# pre-processing--- drop unwanted columns
def pre_process(df):
  df=df.drop(columns=['sno'])
  return df
bnc=pre_process(bnc)
btc=pre_process(btc)
eth=pre_process(eth)
```

```python
# Add coin identifier
bnc["coin"]="Binance"
btc["coin"]="Bitcoin"
eth["coin"]="Ethereum"
```

```python
# Combine all the datasets--- key step
crypto_df=pd.concat([bnc,btc,eth],ignore_index=True)
crypto_df.head()
```

```python
# Convert Date and Sort it
crypto_df["date"]=pd.to_datetime(crypto_df["date"])
crypto_df=crypto_df.sort_values(["coin","date"])
crypto_df.tail()
```

```python
# Create a column for daily returns per coin
crypto_df["daily_return"]=(crypto_df.groupby("coin")["close"].pct_change())
crypto_df.head()
```

```python
crypto_df["daily_return"]=crypto_df["daily_return"].fillna(0)
crypto_df.head()
```

```python
# Calculate volatility or risk ---- tells which coin is riskier ?
# std deviations= measures the spread of data pts around mean
volatility_df=crypto_df.groupby("coin")["daily_return"].std().reset_index().rename(columns={"daily_return":"volatility"})
volatility_df
```

```python
# Average Daily Return--- tells which coin performs better on average ?
avg_return_df=crypto_df.groupby("coin")["daily_return"].mean().reset_index().rename(columns={"daily_return":"avg_return"})
avg_return_df
```

```python
# Now merge risk and return---- gives risk-return table
performance_df=avg_return_df.merge(volatility_df,on="coin")
performance_df
```

```python
# Growth of coin over time--- tells how wealth grows if you invested from day one.
crypto_df["cumulative_return"]=crypto_df.groupby("coin")["daily_return"].transform(lambda x:(1+x).cumprod()-1)
crypto_df.head()
```

```python
# Calculate the running peak--- tells highest cumulative return achieved so far for each coin.
crypto_df["running_peak"]=crypto_df.groupby("coin")["cumulative_return"].transform(lambda x:x.cummax())
crypto_df.head()
```

```python
# Calculate drawdown--- tells how much an asset falls from its previous peak
crypto_df["drawdown"]=(crypto_df["cumulative_return"]-crypto_df["running_peak"])
crypto_df.head()
```

```python
# Max drawdown--- tells what is the worst loss each coin experienced ?
max_drawdown_df =crypto_df.groupby("coin")["drawdown"].min().reset_index().rename(columns={"drawdown": "max_drawdown"})
crypto_df.head()
```

```python
crypto_df.to_csv("crypto_cleaned.csv", index=False)
```
